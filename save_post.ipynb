```python
import instaloader
import os
import json
import pandas as pd
```

```python
L = instaloader.Instaloader(download_videos=False, 
                            download_video_thumbnails=False, 
                            slide="1", 
                            post_metadata_txt_pattern="",
                            save_metadata=False)

# L.login("mbanteam544@gmail.com", "6WijBf*Cw6U4@Qk")

def scrape_data(username):

    profile = instaloader.Profile.from_username(L.context, username)
    
    # create folder

    if not os.path.exists(f"posts_{username}"):
        os.makedirs(f"posts_{username}")

    post_data = pd.DataFrame()
        
    limit = 10 # to be adapted
    posts_ = profile.get_posts()
    i = 0
    for post in posts_:
        if i >= limit:
            break
        
        if not post.is_video:

            # save locally
            L.download_post(post, f"posts_{username}")

            data = {
                "caption": post.caption,
                "likes": post.likes,
                "comments": post.comments,
                "date": post.date.isoformat(),
                "url": post.url
            }

            post_data_i = pd.DataFrame(data, index=range(1))

            post_data = pd.concat([post_data, post_data_i], ignore_index=True)
            
            i += 1
    
    return post_data

def profile_stats(username):
    profile = instaloader.Profile.from_username(L.context, username)

    # store as a json
    data = {
        "username": profile.username,
        "followers": profile.followers,
        "following": profile.followees,
        "bio": profile.biography,
        "posts": profile.mediacount
    }

    profile_data = pd.DataFrame(data, index=range(1))
    
    return profile_data
```

```python
L.login("mbanteam544@gmail.com", "6WijBf*Cw6U4@Qk")

list_username = ["kungfuteausa", "gongchatea", "vivibubbletea", "tenoneteahouse_us", "tigersugar.usa",
                 "superemoji_usa", "0andco"]

post_data = pd.DataFrame()
profile_data = pd.DataFrame()

if __name__ == "__main__":
    for username in list_username:
        post_data_i = scrape_data(username)
        profile_data_i = profile_stats(username)
        
        post_data_i['likes/followers'] = post_data_i['likes']/profile_data_i['followers'].iloc[0]
        post_data_i['comments/followers'] = post_data_i['comments']/profile_data_i['followers'].iloc[0]
        
        post_data_i['username'] = profile_data_i['username'].iloc[0]
        
        post_data = pd.concat([post_data, post_data_i], ignore_index=True)
        profile_data = pd.concat([profile_data, profile_data_i], ignore_index=True)
```
